```python
home = os.path.expanduser('~')
root_dir = 'ds-2155/'
root_dir = os.path.join(home,root_dir)
root_dir
```

```python
files = [file for file in os.listdir(root_dir) if file.startswith('sub-')]
files.sort()
files
```

```python
for f in range(5):
    subfolder = os.path.join(root_dir,files[f],'func') 
    niis = [nii for nii in os.listdir(subfolder) if '.nii' in nii] # grab all scans in a list 
    nii = niis[0] # grab first 
    path_to_scan = os.path.join(subfolder,nii) # make path 
    im = ants.image_read(path_to_scan) # read 4d
    im_3d = ants.slice_image(image=im,axis=3,idx=0) # slice
    im_3d.plot_ortho(flat=True,title=path_to_scan) # plot
```
